# **Business Understanding**
The Public health agency want to improving Public Health Through Vaccine Uptake Prediction.They want to identify high-risk populations who are less likely to get vaccinated and design targeted awareness campaigns to improve uptake.

## **Objective**
The aim is to identify the demographic, behavioral, and attitudinal factors that are most predictive of whether someone receives the H1N1 hence understanding the factors that influence H1N1 vaccination rates and implement them to improve vaccine adoption strategies.
## **Business Questions**
**Demographic Insights:**
 * Do age, income, education level, or employment status impact vaccine uptake?
 * Which demographic groups are least likely to receive the H1N1?

**Behavioral and Attitudinal Factors:**
 * How does perceived risk of infection affect vaccine adoption?
 * Do trust in medical professionals influence vaccine decisions?
 
**Access to Healthcare and Outreach Impact:**
 * Does having health insurance affect vaccination rates?
 * How does awareness impact the likelihood of vaccination?

**Gender impact:**
 * Does gender influence the likelihood of vaccination among the target population?

**Predictive Modeling for Vaccine Hesitancy:**
 * Can we predict which individuals are most likely to refuse the vaccine?

## **Metrics of Success**
**Predictive Modeling for Vaccine Hesitancy:**
 * Model Accuracy: 
    *  When predicting individuals most likely to refuse the vaccine, our primary focus is on recall rather than precision.        
    * Here's why:
       * Recall (Sensitivity/True Positive Rate):Measures how many actual vaccine refusers we correctly identify.
           * A high recall ensures that we capture most of the people who would refuse the vaccine, reducing the chances of                 missing out on them.
           * It is important for public health campaigns since missing a vaccine refuser means lost intervention  opportunities.
       * Precision:Measures how many of our predicted vaccine refusers were actually refusers.
          * High precision means fewer false positives.
          * It's important but less critical because the cost of misclassifying a non-refuser as a refuser is relatively low.  
 * Feature Importance:
    * Analyze which features have the highest impact on predicting vaccine refusal.
 * Confusion Matrix:
    * Use confusion matrix to evaluate the false positives (incorrectly predicting someone will refuse the vaccine) and false negatives (incorrectly predicting someone will accept the vaccine) in the prediction.
 
**Gender Impact:**
 * Use of the Hypothesis:
    * The hypothesis test assess if gender differences are significant and whether interventions should be gender-targeted.

for the remaining key questions good visualizations will be used

### **Missing Values**

```python
#create a function to calculate missing values and their percentage
def missing_values_percentage(df):

    # Total rows 
    total_records = len(df)

    # Missing values 
    missing_values = df.isnull().sum()

    # Percent missing values 
    percent_missing_values = (missing_values/ total_records) * 100

    # # Get the data types of each column
    data_types = df.dtypes  

    # create summary dataframe 
    missing_values_df = pd.DataFrame({
         'missing_values': missing_values,
         'percentage (%)': percent_missing_values,
         'data_type': data_types
    })

    print(f"Total records: {total_records}")
    
    return missing_values_df

# View the dataframe 
missing_summary =missing_values_percentage(combined_df)
missing_summary
```

### **Outliers**
Use interquatile range to identify outliers and if any decide on a strategy to deal with them

```python
num_cols=combined_df.select_dtypes(include=['int64']).columns

# Compute the 25th percentile (Q1) for each numerical column
Q1 = combined_df[num_cols].quantile(0.25)  

# Compute the 75th percentile (Q3) for each numerical column
Q3 = combined_df[num_cols].quantile(0.75)  

# Calculate the Interquartile Range (IQR), which measures the spread of the middle 50% of data
IQR = Q3 - Q1  

# Identify outliers using the IQR rule:
# An outlier is any value below Q1 - 1.5 * IQR (lower bound) or above Q3 + 1.5 * IQR (upper bound)
outliers = ((combined_df[num_cols] < (Q1 - 1.5 * IQR)) |  
            (combined_df[num_cols] > (Q3 + 1.5 * IQR)))

# Count the number of outliers in each numerical column
outlier_counts = outliers.sum()

# Display columns with outliers, sorted from highest to lowest count
print(outlier_counts[outlier_counts > 0].sort_values(ascending=False))
```

```python
#create a map_generation function
def map_generation(age_group):
    if age_group == '18 - 34 Years':
        return 'Gen Z'
    elif age_group =='35 - 44 Years':
        return 'Millenials'
    elif age_group =='45 - 54 Years':
        return 'Gen X'
    elif age_group == '55 - 64 Years':
        return 'Baby Boomers'
    else:
        return 'Silent Generation'

#apply the function
combined_df['generation']=combined_df['age_group'].apply(map_generation)
#view the dataset
combined_df.head()
```

# **Hypothesis Testing**
We conduct a hypothesis test to determine if there is a significant difference in H1N1 vaccination rates between females and males 

## **Define the Hypotheses**
Null Hypothesis ((H_0)): There is no significant difference in the proportion of vaccinated females and males.
* [ H_0: p_f = p_m ] 

Alternative Hypothesis ((H_1)): There is a significant difference in the proportion of vaccinated females and males. 
* [ H_1: p_f != p_m ]

**Where (p_f) is the proportion of vaccinated females and (p_m) is the proportion of vaccinated males.**

## **Collect Data**

You will need the following data:
 * Number of vaccinated females ((X_f))
 * Total number of females surveyed ((n_f))
 * Number of vaccinated males ((X_m))
 * Total number of males surveyed ((n_m))

```python
# Count the number of vaccinated females
X_f = combined_df[(combined_df['sex'] == 'Female') & (combined_df['h1n1_vaccine'] == 1)].shape[0]

# Total number of females surveyed
n_f = combined_df[combined_df['sex'] == 'Female'].shape[0]

# Count the number of vaccinated males
X_m = combined_df[(combined_df['sex'] == 'Male') & (combined_df['h1n1_vaccine'] == 1)].shape[0]

# Total number of males surveyed
n_m = combined_df[combined_df['sex'] == 'Male'].shape[0]

# Display results
print(f"Number of vaccinated females (X_f): {X_f}")
print(f"Total number of females surveyed (n_f): {n_f}")
print(f"Number of vaccinated males (X_m): {X_m}")
print(f"Total number of males surveyed (n_m): {n_m}")
```

## **Calculate Sample Proportions**
Calculate the sample proportions for both groups:


**Sample proportion of vaccinated females:**  
$$\hat{p_f} = \frac{X_f}{n_f}$$

**Sample proportion of vaccinated males:**  
$$\hat{p_m} = \frac{X_m}{n_m}$$


```python
# Calculate sample proportions
p_f_hat = X_f / n_f
p_m_hat = X_m / n_m

# Display results
print(f"Sample proportion of vaccinated females: {p_f_hat}")
print(f"Sample proportion of vaccinated males: {p_m_hat}")
```

## **Calculate the Standard Error (SE)**
The standard error (SE) for the difference in proportions is calculated as:  

$$
SE = \sqrt{\hat{p_f}(1 - \hat{p_f}) \frac{1}{n_f} + \hat{p_m}(1 - \hat{p_m}) \frac{1}{n_m}}
$$


```python
# Calculate the Standard Error (SE)
SE = np.sqrt(p_f_hat * (1 - p_f_hat) / n_f + p_m_hat * (1 - p_m_hat) / n_m)
# Display results
print(f"standard error (SE) : {SE}")
```

## **Calculate the Z-Statistic**
The z-statistic for the difference in proportions is calculated as:
    
$$
z = \frac{\hat{p_f} - \hat{p_m}}{SE}
$$

```python
# Calculate the Z-Statistic
z = (p_f_hat - p_m_hat) / SE
# Display results
print(f" z-statistic: {z}")
```

## **Determine the Critical Value**
In a two-tailed test with a significance level of $ \alpha = 0.05 $, the critical z-values are approximately $ \pm 1.96 $, meaning any z-score beyond this range indicates statistical significance.



```python
#import library
import scipy.stats as stats

# Determine the critical value for a two-tailed test at alpha = 0.05
alpha = 0.05
critical_value = stats.norm.ppf(1 - alpha / 2)

print(f"Critical Value: {critical_value:.4f}")
```

## **Make a Decision**
If the absolute value of the calculated z-statistic is greater than the critical value, you reject the null hypothesis.

```python
# Compare Z-Statistic with Critical Value
if abs(z) > critical_value:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")

# Conclusion
if abs(z) > critical_value:
    print("Conclusion: There is a statistically significant difference in the proportion of vaccinated females and males.")
else:
    print("Conclusion: There is no statistically significant difference in the proportion of vaccinated females and males.")
```

 ## Encoding
This is the process of transforming categorical variables into numerical formats that can be used when modelling.We will use binary encoding on columns that contain only two types of categorical data and onehotencoding on the other categoricalcal columns 

### creating a pipeline for encoding
what is a pipeline?A pipeline in machine learning is a structured approach to automate and streamline the workflow of data processing, model training, and evaluation. It enhances reproducibility, modularity, and efficiency, making it an essential tool in data science and machine learning projects.In Python, particularly with libraries like scikit-learn, you can create a pipeline using the Pipeline class.


```python
#import libraries
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, FunctionTransformer
```

```python
# Define columns
binary_columns = ['sex', 'marital_status', 'Housing_situation']
onehot_columns = ['race', 'employment_status', 'generation', 'education_lvl', 'income_category']
integer_columns = [
    'h1n1_concern_lvl', 'h1n1_knowledge', 'taken_antiviral_meds', 'infected_avoidance', 'bought_face_mask',
    'wash_hands', 'less_large_gatherings', 'less_contact_outside_home', 'avoid_face_touching', 'doctor_recc_h1n1',
    'chronic_med_condition', 'contact_with_infant', 'health_worker', 'health_insurance', 'h1n1_vaccine_effectiveness_op',
    'h1n1_risk_perception', 'h1n1_vaccine_safety_concern', 'other_household_adults', 'household_children', 'h1n1_vaccine',
    'vaccine_confidence', 'awareness_score', 'medical_trust_score', 'family_responsibility_score'
]

# Function to convert categorical values to binary
def binary_encode(X):
    return X.replace({'Male': 1, 'Female': 0, 'Married': 1, 'Not Married': 0, 'Rent': 1, 'Own': 0}).fillna(0)

# Define the ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('binary', FunctionTransformer(binary_encode), binary_columns),
        ('onehot', OneHotEncoder(drop='first', sparse_output=False), onehot_columns)
    ], remainder='passthrough'  # Keep integer columns unchanged
)

# Create pipeline
pipeline = Pipeline([
    ('preprocessor', preprocessor)# pipeline ensures that the transformations happen sequentially
])

# Fit and transform the data
transform_data = pipeline.fit_transform(combined_df)

# Get column names for one-hot encoding features
onehot_columns_names = pipeline.named_steps['preprocessor'].transformers_[1][1].get_feature_names_out(onehot_columns)

# Combine all feature names
all_columns = integer_columns+binary_columns + list(onehot_columns_names)

# Create the final DataFrame
transformed_df = pd.DataFrame(transform_data, columns=all_columns)

# View the encoded dataset
transformed_df.head()
```

## Hyperparameter search using optuna 
Optuna is an automatic hyperparameter optimization framework that efficiently searches for the best hyperparameters using adaptive sampling and pruning techniques.
Why Use Optuna?:
 * Automated Hyperparameter Tuning: Finds the best model parameters efficiently.
 * Tree-structured Parzen Estimator (TPE): Uses a smarter Bayesian optimization approach.
 * Pruning: Stops unpromising trials early to save time.
 * Parallel Execution
 * Easy Integration.

### Define the Optuna Optimization Function

```python
import optuna
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.metrics import roc_auc_score, make_scorer

import optuna
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.metrics import recall_score, make_scorer
import numpy as np

# Custom scorer to get mean recall for both classes
def balanced_recall_scorer(y_true, y_pred_smote):
    recall_per_class = recall_score(y_true, y_pred_smote, average=None)  # Get recall for both classes
    return np.mean(recall_per_class)  # Optimize mean recall for class 0 & 1

def objective(trial):
    # Suggest hyperparameters for Random Forest
    n_estimators = trial.suggest_int("n_estimators", 50, 300, step=50)
    max_depth = trial.suggest_int("max_depth", 3, 20)
    min_samples_split = trial.suggest_int("min_samples_split", 2, 10)
    
    # Define the model
    model = RandomForestClassifier(
        n_estimators=n_estimators, 
        max_depth=max_depth, 
        min_samples_split=min_samples_split, 
        random_state=42
    )
    
    # Perform cross-validation using custom recall scorer
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    recall = cross_val_score(model,X_train_smote, y_train_smote, cv=cv, scoring=make_scorer(balanced_recall_scorer)).mean()
    
    return recall  # Optimizing recall for both classes
```

### Run Optuna Optimization

```python
# Run Optuna optimization
study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=10)

# Best hyperparameters
print("Best hyperparameters:", study.best_params)
print("Best balanced recall score:", study.best_value)
```

### Train Random Forest with Best Parameters

```python
# Retrieve the best hyperparameters
best_params = study.best_params

# Train the final RandomForestClassifier
best_rf_model = RandomForestClassifier(
    **best_params,
    random_state=42,
    n_jobs=-1
)
best_rf_model.fit(X_train_smote, y_train_smote)

# Predict on the test set
y_pred_best = best_rf_model.predict(X_test)
y_pred_proba = best_rf_model.predict_proba(X_test)[:, 1]

# Evaluate performance
print("\n Classification Report:")
print(classification_report(y_test, y_pred_best))
# Print the confusion matrix
print('Confusion Matrix on testing Set:')
print(confusion_matrix(y_test, y_pred_best))
```

After using optuna our recall for class 0 decreased slightly decreased.
